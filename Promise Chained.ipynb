#Collecting Code snippset together

```python
#Importing required library
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('brown')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
import pandas as pd

import locationtagger
import nltk
nltk.download('words') 

import re
#Elimination of Stop words
from nltk.corpus import stopwords 
from nltk.util import ngrams 

#Counting the frequency
from collections import Counter

#NLP Toolkit
import nltk

# For visualization:
import matplotlib.pyplot as plt
import seaborn as sns


#Functions for future uses
def listToString(s):
    # initialize an empty string
    str1 = " "
    # return string 
    return (str1.join(s))

#Function to remove the sentences starting with particular character or pattern
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt   

#Extracting id from the details
def can_convert_to_int(string):
    try:
        int(string)

        return True
    except ValueError:
        return False


#Importing easyocr library for Optical character recognition that would be further used in Text Extraction from Documents. example Govermental Proofs.
import easyocr
reader = easyocr.Reader(['en']) # this needs to run only once to load the model into memory

image=input("Kindly paste the address of Image: ")

#Taking Scanned image as input and reading text from it using OCR
result = reader.readtext(image)

#Saving text 
Data=reader.readtext(image, detail = 0)


#Implementing NLP
import numpy as np
#removing hashtags (word starting with #)
Data= np.vectorize(remove_pattern)(Data, "#[\w]*")

#removing hashtags (word starting with https)
Data= np.vectorize(remove_pattern)(Data, "https[\w]*")

#remove user, https, and RT
Data = np.vectorize(remove_pattern)(Data, "https|RT|@[\w]*")

#remove stop words
stop_words = set(stopwords.words('english')) 

#Printing the Data of the document
details=" "
for ele in Data:
  details=details+ " "+ ele


#Extracting details from the text
place_entity = locationtagger.find_locations(text = details)

 
# # getting all countries
# print("The countries in text : ")
# print(place_entity.countries)
 
# getting all states
print("The states in text : ")
place=place_entity.regions
print(place_entity.regions)

id=""
for ele in Data:
  if len(ele)==6:
    if can_convert_to_int(ele)==True:
      id=ele
      print("The is is: ", id)
      break
 
# Import writer class from csv module
from csv import writer

# List that we want to add as a new row
List = [id,place,details]

# Open our existing CSV file in append mode
# Create a file object for this file
with open('/content/drive/MyDrive/Hackathon_Devfolio/Details.csv', 'a') as f_object:

	# Pass this file object to csv.writer()
	# and get a writer object
	writer_object = writer(f_object)

	# Pass the list as an argument into
	# the writerow()
	writer_object.writerow(List)

	# Close the file object
	f_object.close()


```
